```python
import numpy as np
```

```python
data = df = pd.read_csv('weather12.csv')
df = pd.DataFrame(data)
```

```python
df.head(5)
```

```python
def calculate_entropy(column):
  unique_values, counts = np.unique(column, return_counts=True)
  probabilities = counts / len(column)
  entropy = -np.sum(probabilities * np.log2(probabilities))
  return entropy
```

```python
entropies = {}

for column in df.columns:
  entropies[column] = calculate_entropy(df[column])

print("entropy for each column:")
for column, entropy in entropies.items():
  print(f"{column}: {entropy}")

parent_entropy = calculate_entropy(df['RainTomorrow'])
```

```python
def information_gain(parent_entropy, feature_column):
  weighted_values = np.unique(feature_column)
  weighted_entropy = 0
  for value in unique_values:
    subset = df[df[feature_column.name]  == value]
    subset_entropy = calculate_entropy(subset['RainTomorrow'])
    weight = len(subset) / len(df)

    weighted_entropy += weight * subset_entropy
    info_gain = parent_entropy - weighted_entropy
    return info_gain
```

```python
information_gains = {}

for column in df.drop(columns=['RainTomorrow']).columns:
    information_gains[column] = information_gain(parent_entropy, df[column])

print("\nInformation gain for each feature")

for column, gain in information_gains.items():
  print(f"{column}: {gain}")

selected_feature = max(information_gains, key=information_gains.get)
print("\n selected feature :" , selected_feature)
```

```python
x = df[[selected_feature]]
y = df['RainTomorrow']

clf = DecisionTreeClassifier()
clf.fit(x,y)


plot_tree(clf, feature_names = [selected_feature], class_names=['No', 'Yes'], filled =True)
```
