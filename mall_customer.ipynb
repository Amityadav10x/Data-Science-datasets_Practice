```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mydata = pd.read_csv("Mall_Customers.csv")
```

```python
mydata.head(5)
```

```python
mydata.tail(5)
```

```python
iv = mydata[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2,n_init=10, random_state=0)
kmeans.fit(iv)
clusters = kmeans.predict(iv)
```

```python
wss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i, n_init=10, random_state=0)
  kmeans.fit(iv)
  wss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,  11), wss)
plt.title('the Elbow Plot')
plt.xlabel("Number of clusters")
plt.ylabel('within sum of square')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, n_init=10, random_state=0)
clusters = kmeans.fit_predict(iv)
```

```python
iv['cluster'] = clusters
plt.scatter(iv.loc[iv['cluster'] == 0, 'Annual Income (k$)'],
            iv.loc[iv['cluster'] == 0, 'Spending Score (1-100)'], s=100, c='red', label='Careful')
plt.scatter(iv.loc[iv['cluster'] == 1, 'Annual Income (k$)'],
            iv.loc[iv['cluster'] == 1, 'Spending Score (1-100)'], s=100, c='green', label='Standard')
plt.scatter(iv.loc[iv['cluster'] == 2, 'Annual Income (k$)'],
            iv.loc[iv['cluster'] == 2, 'Spending Score (1-100)'], s=100, c='blue', label='Target')
plt.scatter(iv.loc[iv['cluster'] == 3, 'Annual Income (k$)'],
            iv.loc[iv['cluster'] == 3, 'Spending Score (1-100)'], s=100, c='grey', label='Careless')
plt.scatter(iv.loc[iv['cluster'] == 4, 'Annual Income (k$)'],
            iv.loc[iv['cluster'] == 4, 'Spending Score (1-100)'], s=100, c='brown', label='Sensible')




plt.title("Results of K means clustering")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Mall_Customers.csv")
```

```python
data.head(5)
```

```python
data = pd.DataFrame({
    'Annual Income (k$)': [15,15,16,16,17,17,18,18,19,19,20,20,21,21],
    'Spending Score (1-100)':[39,81,6,77,40,76,6,94,3,72,14,99,15,77]
})
```

```python
mean_values = np.mean(data, axis=0)
std_dev = np.std(data, axis=0)
standardized_data = (data - mean_values) / std_dev
```

```python
cov_matrix = np.cov(standardized_data, rowvar = False)
```

```python
eigen_values, eigen_vectors = np.linalg.eig(cov_matrix)



plt.bar(np.arange(len(eigen_values)), eigen_values)
plt.xlabel('pRICIPLE cOmponent')
plt.ylabel('Eigenvalue')
plt.xlabel('Eigenvector')
plt.show()
```

```python
plt.plot(eigen_vectors[:, 0], label = 'Eigenvector 1')
plt.plot(eigen_vectors[:, 1], label = 'Eigenvector 2')
plt.xlabel('Feature')
plt.ylabel('value')
plt.legend()
plt.show()
```

```python
plt.imshow(cov_matrix, cmap='hot',interpolation='nearest' )
plt.title('Covariance Matrix')
plt.colorbar()
plt.show()
```
